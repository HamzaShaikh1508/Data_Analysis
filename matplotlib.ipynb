```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\hamza_sk\\Downloads\\train.csv", parse_dates=['Order Date','Ship Date'])
del df['Order ID']
del df['Row ID']
df.head()
```

```python
df['Year'] = df['Order Date'].dt.year
xdf = df.groupby('Year')['Sales'].sum()
xdf
```

```python
ax = xdf.plot(kind = 'bar')
ax.grid()
```

```python
ax = xdf.plot(kind = 'bar', color = 'g', alpha = 0.2)
```

```python
xdf.plot()
```

```python
ax = xdf.plot()
ax.set_xticks(xdf.index)
```

```python
ax = xdf.plot()
ax.set_xticks(xdf.index)
ax.set_ylabel("Sales")
ax.set_title("Yearly Sales")
ax
```

```python
ax = xdf.plot(linestyle = 'dashed')
ax.set_xticks(xdf.index)
ax.set_ylabel("Sales")
ax.set_title("Yearly Sales")
ax
```

```python
ax = xdf.plot(linestyle = 'dotted')
ax.set_xticks(xdf.index)
ax.set_ylabel("Sales")
ax.set_title("Yearly Sales")
ax
```

```python
ax = xdf.plot(linestyle = 'dashdot')
ax.set_xticks(xdf.index)
ax.set_ylabel("Sales")
ax.set_title("Yearly Sales")
ax
```

```python
ax = xdf.plot(linewidth = 3, color = 'g', linestyle = 'dotted', alpha = 0.2)
ax.set_xticks(xdf.index)
ax.set_ylabel("Sales")
ax.set_title("Yearly Sales")
ax
```
